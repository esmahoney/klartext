#KlarText Overview

In the collection of notebooks we are exploring baseline and how to move forward with strategy. The idea is not to "beat big LLMs" but rather prove something product shaped (poc) that big generic models can't provide out of the box.

With the baseline we are trying to that we can deliver a predictable, safe, accessible simplication experience for EN/DE at an acceptable cost with a clear upgrade path. Using big models only when necessary.

1. Reliablity claim:
We can simplify without breaking critical facts.

Success metric: Numbers/dates preserved rate (target: >98% on "golden set")

Failure metric: “meaning-breaking” changes (negations, obligations: must/may/shall) rate (target: near 0)

This is where generic LLMs often fail unless you build guardrails. We want to:
* guarentee numbers/dates/negations are preserved
* provide consistent style control across EN and DE
* provide accessibility focused formatting (predictable bullets, headings, step-by-step, short sentences)
* auditability - why this output, what changed, what checks passed
* cost discipline (routing and caching)


2. Accessibility claim:
Outputs are materially easier to read

Metrics: simplicity score + readability proxies (sentence length, jargon rate)

Human rubric: "Would your target user understand it better?" (target: majority "yes")

These outputs follow specific rules: 
Sentence Level
* A sentence should not be longer than 15 - 20 words
* Ideally, a sentence should contain only one comma.
* Avoid parenthetical insertions and nested (complex) sentences.
* Use active verbs; avoid the passive voice.
* Write clear, unambiguous sentences.

Word level
* Avoid foreign words whenever possible.
* Explain difficult and long (compound) words.
* You can write long or compound words with a hyphen.
* Avoid idioms, irony, and metaphors.
* Write out abbreviations whenever possible.
* Avoid negations.
* Avoid synonyms and stick to one term.

3. Consistency claim:
Same input + same settings = same style

Metric: variance across runs and across models

Product value: “Plain” vs “Extra simple” behaves predictably


4. Cost/control claim
KlarText will cover most requests locally and only pay for API when needed

Metric: % routed to local vs API (target: e.g., 60–80% local for low-risk cases)
Metric: average cost per request (target: within your budget)
Metric: latency (target: reasonable, e.g., <2–4s typical for MVP)

5. Domain fit claim
KlarText handles the kinds of texts that users actually paste (web, admin, health) better than a naïve “just prompt GPT” demo.
Metric: fewer catastrophic errors
Metric: better structure (bullets when helpful, not random)
